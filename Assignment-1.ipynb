## 2. Create the below pattern using nested for loop in Python.


```python
top_star=5
for x in range(1,top_star+1):
    for y in range(0,x):
        print("*", end =" ")
    print()
for x in range(top_star-1,0,-1):
    for y in range(0,x):
        print("*", end =" ")
    print()
```
